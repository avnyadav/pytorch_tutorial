```python
import numpy as np
```

```python
def sigmoid(x):
    return 1/(1+np.exp(x))
```

```python
def d_sigmoid(x):
    return sigmoid(x)*(1-sigmoid(x))
```

```python
def forward(input_feature,weight):
    prediction=sigmoid(np.dot(input_feature,weight))
    return prediction
```

```python
def loss(ytrue,ypred):
    postive_class=ytrue*np.log(ypred)
    negative_class=(1-ytrue)*np.log(1-ypred)
    return -1*np.sum((postive_class+negative_class)/len(ytrue))
```

```python
def backword_prop(target_column,y_pred):
    postive=np.sum((target_column*y_pred*(1-y_pred))/(np.log(y_pred)))/len(y_pred)
    negative=-1*np.sum(((1-target_column)*y_pred*(1-y_pred)/np.log(1-y_pred))/len(y_pred))
    return postive+negative
    
```

```python
def get_new_weight(current_weight,y_true,y_pred,lr=0.01):
    return current_weight-(lr*backword_prop(y_true,y_pred))
```
