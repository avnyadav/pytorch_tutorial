```python
def relu(x):
    return (x>0).astype(int)*x
```

```python
def d_relu(x):
    return (x>0).astype(int)
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.arange(-2,2,0.01)
```

```python
plt.plot(x,relu(x),label="relu")
plt.plot(relu(x),d_relu(relu(x)),label="derivative")
plt.legend()
plt.show()
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def d_sigmoid(x):
    return sigmoid(x)*(1-sigmoid(x))
```

```python
x=np.arange(-10,10,0.001)
```

```python
plt.plot(x,sigmoid(x),label="Sigmoid function")
plt.plot(x,d_sigmoid(x),label="Derivative of Sigmoid ")
plt.legend()
```

```python
def tanh(x):
    return (2*sigmoid(2*x))-1
```

```python
def d_tanh(x):
    return 1-tanh(x)**2
```

```python
plt.plot(x,tanh(x),label="tanh")
plt.plot(x,d_tanh(x),label="derivative of tanh")
plt.show()
```

```python
def softmax(x):
    exp_x=np.exp(x-x.max())
    return exp_x/np.sum(exp_x)
```

```python
softmax(np.array([5,6,7]))
```

```python
def softplus(x):
    return np.log(1+exp(x))
```
